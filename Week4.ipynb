```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

img_f = cv.imread('/content/drive/MyDrive/PCVK2023/Images/female.tiff')
img_rgb = cv.cvtColor(img_f,cv.COLOR_BGR2RGB) # konversi channel BGR -> RGB
plt.imshow(img_rgb)
```

```python
# 1 Buat Gamma Correction
print(' Gamma Correction pada citra ')
print(' --------------------------- ')
try:
  gamma = int(input(' Masukkan nilai Gamma :'))
except ValueError:
  print('Error, not a number')

original = cv.cvtColor(img_rgb,cv.COLOR_BGR2RGB)
gamma_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
     gamma_image[y,x,c] = np.clip(255 * pow(original[y,x,c] / 255, (1/gamma)), 0, 255)

final_frame = cv.hconcat((original, gamma_image))
cv2_imshow(final_frame)
```

```python
# 2 Simulasi Image Depth
print(' Simulasi Image Depth ')
print('----------------------')

original = cv.imread('/content/drive/MyDrive/PCVK2023/Images/female.tiff', cv.IMREAD_GRAYSCALE)
img_depth = np.zeros(original.shape, original.dtype)

# nilai kedalaman
depth=2

# Rumus menentukan level kedalaman
level = 255 / ((2**depth) - 1)

# Akses gambar tiap pixel
for y in range(original.shape[0]):
  # Rumus Bit Depth
  img_depth[y] = np.clip(np.round(original[y]/level)*level, 0, 255)

final_frame = cv.hconcat((original, img_depth))
cv2_imshow(final_frame)
```

```python
# 3 Buat modul Average Denoising (Citra 5)
import glob
from math import log10, sqrt

original = cv.imread('/content/drive/MyDrive/PCVK2023/Images/galaxy.jpg')
cv_img = []
for img in glob.glob('/content/drive/MyDrive/PCVK2023/Images/noises/*.jpg'):
    n = cv.imread(img)
    cv_img.append(n)

dst = cv_img[0]
noise_number = 4 # nomor (array) gambar pada folder noises
for i in range(noise_number):
    if i == 0:
        pass
    else:
        alpha = 1.0 / (i + 1)
        beta = 1.0 - alpha
        dst = cv.addWeighted(cv_img[i], alpha, dst, beta, 0.0)

def PSNR(original, compressed):
    mse = np.mean((original - compressed) ** 2)
    if(mse == 0):
        return 100, mse
    max_pixel = 255.0
    psnr = 20 * log10(max_pixel / sqrt(mse))
    return psnr, mse


compressed = dst
psnr = PSNR(original, compressed)
galaxy_frame = cv.hconcat((original, dst))
cv2_imshow(galaxy_frame)


psnr, mse = PSNR(original, compressed)
print('Nilai PSNR adalah : ', psnr, '(dB)')
print('Nilai MSE adalah  : ', mse)
```

```python
# Tugas 4
# Buat image masking untuk image berikut. Image kiri adalah image asli (images/couple.tiff),
# sedangkan image paling kanan adalah hasilnya:
img_rose = cv.imread('/content/drive/MyDrive/PCVK2023/Images/couple.tiff')
mask = np.zeros(img_rose.shape, dtype=np.uint8)
mask = cv.circle(mask, (70, 40), 40, (255,255,255), -1)
mask1 = cv.circle(mask, (195, 40), 40, (255,255,255), -1)


result = cv.bitwise_and(img_rose, mask)

result[mask==0] = 255

final_frame = cv.hconcat((img_rose, mask, result))
cv2_imshow(final_frame)
```

```python
# Tugas 4
# Buat image masking untuk image berikut Image kiri adalah image asli (images/teeth.jpg) Sedangkan image kanan adalah hasilnya

img_teeth = cv.imread('/content/drive/MyDrive/PCVK2023/Images/teeth.jpg')

mask = np.zeros(img_teeth.shape, dtype=np.uint8)
mask = cv.rectangle(mask,pt1=(315,595),pt2=(770,1025),color=(255,255,255),thickness=-1)

result = img_teeth & mask

mask1 = cv.rectangle(mask,pt1=(405,5),pt2=(850,455),color=(255,255,255),thickness=-1)

result = img_teeth & mask1

teeth_frame = cv.hconcat((img_teeth, result))
cv2_imshow(teeth_frame)
```

```python
# Tugas 5
# Lakukan percobaan menggunakan operator lain dan tunjukkan hasilnya pada modul ini

img_rose = cv.imread('/content/drive/MyDrive/PCVK2023/Images/rose_pink.png')
```

```python
# OR
mask_rose = np.zeros(img_rose.shape, dtype=np.uint8)
mask_rose = cv.circle(mask_rose, (260, 300), 225, (255,255,255), -1)

masked = img_rose | mask_rose

rose_frame = cv.hconcat((img_rose, mask_rose, masked))
cv2_imshow(rose_frame)
```

```python
# NOT
mask_rose = np.zeros(img_rose.shape, dtype=np.uint8)
mask_rose = cv.circle(mask_rose, (260, 300), 225, (255,255,255), -1)

masked = ~img_rose & ~mask_rose

rose_frame = cv.hconcat((img_rose, mask_rose, masked))
cv2_imshow(rose_frame)
```

```python
# NAND

mask_rose = np.zeros(img_rose.shape, dtype=np.uint8)
mask_rose = cv.circle(mask_rose, (260, 300), 225, (255,255,255), -1)

masked = ~img_rose & mask_rose

rose_frame = cv.hconcat((img_rose, mask_rose, masked))
cv2_imshow(rose_frame)
```

```python
# XOR

mask_rose = np.zeros(img_rose.shape, dtype=np.uint8)
mask_rose = cv.circle(mask_rose, (260, 300), 225, (255,255,255), -1)

masked = img_rose ^ mask_rose

rose_frame = cv.hconcat((img_rose, mask_rose, masked))
cv2_imshow(rose_frame)
```

```python
# AND

mask_rose = np.zeros(img_rose.shape, dtype=np.uint8)
mask_rose = cv.circle(mask_rose, (260, 300), 225, (255,255,255), -1)

masked = img_rose & mask_rose

rose_frame = cv.hconcat((img_rose, mask_rose, masked))
cv2_imshow(rose_frame)
```
